```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
#all_players consist of the data for every player from each season in the range 1996-1997 to 2019-2020 
all_players = pd.read_csv('all_seasons.csv')
```

```python
all_players
```

```python
#Interested in players that were drafted within the top 3 (now known as lottery picks)
picks = ["1","2","3"]
top_3_picks = all_players[all_players['draft_number'].isin(picks)]
```

```python
top_3_picks
```

```python
#Interested in solely the rookie year of each top 3 draft pick
top_3_picks_rookie_year = top_3_picks.drop_duplicates('player_name', keep='first')
```

```python
top_3_picks_rookie_year
```

```python
#Target the rookie years of 1996 - 2019 as the data set begins with the 1996-1997 season but includes players that were drafted previously but still played in this season
years = ['1996','1997','1998','1999','2000','2001','2002','2003','2004','2005','2006','2007','2008','2009','2010','2011','2012','2013','2014','2015','2016','2017','2018','2019']
top_3_picks_final = top_3_picks_rookie_year[top_3_picks_rookie_year['draft_year'].isin(years)]
```

```python
year_2003 = top_3_picks_final[top_3_picks_final['draft_year'] == '2003']
```

```python
year_2003
```

```python
#Chart to breakdown the spread of Rookie season PPG performances for the top 3 picks since 1996
#Potential for further study is fitting a line to the pattern to predict the rookie season PPG for the next 3 draft picks
#This of course is not generalizable as there are too many factors not being controlled or accounted for 
#such as whether the individual is starting or coming off the bench, what team they're playing for, college statistics, etc.
plt.figure(figsize=(15,4))
plt.plot(top_3_picks_final['draft_year'],top_3_picks_final['pts'],'o')
plt.xticks(rotation=55)
plt.title("Avg Rookie season PPG of NBA draft picks")
plt.xlabel("Year")
plt.ylabel("Average PPG")
plt.show()
```

```python
plt.figure(figsize=(15,4))
plt.plot(top_3_picks_final['draft_year'],top_3_picks_final['reb'],'o')
plt.xticks(rotation=55)
plt.title("Avg Rookie season RPG of NBA draft picks")
plt.xlabel("Year")
plt.ylabel("Average RPG")
plt.show()
```

```python
plt.figure(figsize=(15,4))
plt.plot(top_3_picks_final['draft_year'],top_3_picks_final['ast'], 'o')
plt.xticks(rotation=55)
plt.title("Avg Rookie season APG of NBA draft picks")
plt.xlabel("Year")
plt.ylabel("Average APG")
plt.show()
```
