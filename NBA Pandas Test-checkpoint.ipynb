```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
#all_players consist of the data for every player from each season in the range 1996-1997 to 2019-2020 
source = pd.read_csv('all_seasons.csv')
source
```

```python
#Create a new data frame to capture the specific data sets we're interested in
all_players_one = source[['player_name','draft_year', 'draft_round', 'pts','reb','ast']].copy(deep=True)

#Focusing on the average ppg, rpg, and apg we will remove any individuals that went undrafted in a year's draft.
all_players_one = all_players_one[~all_players_one.draft_year.str.contains("Undrafted")]
all_players_one
```

```python
#We now need to validate the data types of the values we have now captures in dataframe 'all_players'
all_players_one.dtypes
```

```python
#We want to change 'draft_year' to value type int
all_players = all_players_one[['player_name', 'pts', 'reb', 'ast']]
all_players.insert(1, 'draft_year', pd.to_numeric(all_players_one['draft_year']), True)
```

```python
all_players.dtypes
```

```python
all_players
```

```python
#Limit the data set to the first time we see a player's name. This will be the player's rookie season or the first time we see
#a player starting from the 1996-97 season
first_entry = all_players.drop_duplicates('player_name', keep='first')
```

```python
first_entry
```

```python
#Target the rookie classes of the 1996 - 2019 drafts. As 'draft_year' is now of type int, we can apply this filter to the values
rookie_seasons = first_entry[first_entry['draft_year'] > 1995]
```

```python
rookie_seasons
```

```python
#Creating a new dataframe to contain the average points per game
agg_df_ppg = rookie_seasons.groupby('draft_year', as_index=False)['pts'].mean().rename(columns={'pts': 'avg_ppg'})
rookie_seasons_ppg = rookie_seasons.merge(agg_df_ppg)
rookie_seasons_ppg
```

```python
#Creating a new dataframe to contain the average rebound per game
agg_df_rpg = rookie_seasons.groupby('draft_year', as_index=False)['reb'].mean().rename(columns={'reb': 'avg_rpg'})
rookie_seasons_ppg_rpg = rookie_seasons.merge(agg_df_rpg)
rookie_seasons_ppg_rpg
```

```python
#Creating a new dataframe to contain the average assists per game
agg_df_apg = rookie_seasons.groupby('draft_year', as_index=False)['ast'].mean().rename(columns={'ast': 'avg_apg'})
rookie_seasons_ppg_rpg_apg = rookie_seasons_ppg_rpg.merge(agg_df_apg)
rookie_seasons_ppg_rpg_apg
```

```python
#Creating a new dataframe to isolate the values Draft_year, avg_ppg, avg_rpg, and avg_apg to then remove duplicates
final = rookie_seasons_ppg_rpg_apg[['draft_year','avg_ppg','avg_rpg','avg_apg']]
#Establishing the final dataframe that removes duplicates so each year has the rookie draft class' statlines 
final_df = final.drop_duplicates('draft_year', keep='first')
final_df
```

```python
#Scatterplot to document the average statline for each draft class' rookie season since 1996. 
#This of course is not generalizable as there are too many factors not being controlled or accounted for 
#such as whether the individual is starting or coming off the bench, what team they're playing for, college statistics, etc.
plt.figure(figsize=(15,4))
plt.plot(final_df['draft_year'],final_df['avg_ppg'],'-o',label="PPG")
plt.plot(final_df['draft_year'],final_df['avg_apg'], '-o',label="APG")
plt.plot(final_df['draft_year'],final_df['avg_rpg'],'-o',label="RPG")
plt.legend()
plt.xticks(rotation=55)
plt.title("Each draft class' Avg Rookie season statline")
plt.xlabel("Year")
plt.ylabel("Average")
plt.show()
```
